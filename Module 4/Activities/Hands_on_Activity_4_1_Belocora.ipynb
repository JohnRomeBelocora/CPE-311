#Problem:

1. Only one disk can be transfer at a time.
2. Each move consists of taking the upper disk from one of the peg and placing it on the top of another peg i.e. a disk can only be moved if it is the uppermost disk of the peg.
3. Never a larger disk is placed on a smaller disk during the transfer.




image.png

```python
def TowerOfHanoi(n , source, destination, auxiliary):
    if n==1: #determines if the inputted value is equal to 1
        print ("Moved disk 1 from source",source,"to destination",destination) #prints if the condition is satisfied
        return #returns the value
    else:
      TowerOfHanoi(n-1, source, auxiliary, destination) #calls out the function again and deducts the n's value with 1
      print ("Moved disk",n,"from source",source,"to destination",destination) #prints the move from source to destination
      TowerOfHanoi(n-1, auxiliary, destination, source) #switches the position of source to auxilary


n = int(input('Enter the number of disks: ')) #asks for user input then gives the value for variable n
TowerOfHanoi(n,'A','B','C') #gives variables for source, destination, and auxilary
```
